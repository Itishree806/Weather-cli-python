```python
import requests
from google.colab import userdata

API_KEY = userdata.get('WEATHER_API_KEY')
BASE_URL = "https://api.weatherapi.com/v1/current.json"

def get_weather(city):
    params = {
        "key": API_KEY,
        "q": city + ", India",
        "aqi": "yes"
    }

    response = requests.get(BASE_URL, params=params)

    if response.status_code == 200:
        data = response.json()

        location  = data["location"]["name"]
        country   = data["location"]["country"]
        temp_c    = data["current"]["temp_c"]
        temp_f    = data["current"]["temp_f"]
        condition = data["current"]["condition"]["text"]
        humidity  = data["current"]["humidity"]
        wind_kph  = data["current"]["wind_kph"]

        aqi       = data["current"]["air_quality"]
        co        = aqi["co"]
        pm2_5     = aqi["pm2_5"]
        pm10      = aqi["pm10"]
        us_index  = aqi["us-epa-index"]

        aqi_labels = {
            1: "Good 😊",
            2: "Moderate 🙂",
            3: "Unhealthy for Sensitive Groups 😷",
            4: "Unhealthy 😨",
            5: "Very Unhealthy 🤢",
            6: "Hazardous ☠️"
        }
        aqi_status = aqi_labels.get(us_index, "Unknown")

        print(f"\n📍 Weather in {location}, {country}")
        print(f"🌡️  Temperature : {temp_c}°C / {temp_f}°F")
        print(f"🌤️  Condition   : {condition}")
        print(f"💧 Humidity    : {humidity}%")
        print(f"💨 Wind Speed  : {wind_kph} kph")
        print(f"\n🌫️  Air Quality")
        print(f"   CO          : {co:.2f} μg/m³")
        print(f"   PM2.5       : {pm2_5:.2f} μg/m³")
        print(f"   PM10        : {pm10:.2f} μg/m³")
        print(f"   AQI Status  : {aqi_status}\n")

    else:
        print(f"❌ Error {response.status_code} - City not found or invalid request!")

if __name__ == "__main__":
    city = input("Enter city name: ")
    get_weather(city)
```
